# BIRCH - Balanced Iterative Reducing and Clustering Hierarchies

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
from sklearn.metrics import silhouette_score
```

### Generate Synthetic Data (3 clusters)

```python
X, y = make_blobs(n_samples = 1000, centers = 3, cluster_std = 0.60,
                 random_state = 0)
```

### Apply BIRCH Clustering

```python
birch = Birch(n_clusters = 3)
y_birch = birch.fit_predict(X)
```

### Evaluate the Clustering Performance using Silhouette Score

```python
silhouette = silhouette_score(X, y_birch)
print(f"Silhouette Score: {silhouette:.4f}")
```

### Visualiz the clustering results

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_birch, cmap='viridis', edgecolors='k', s=50)
plt.title(f'BIRCH Clustering (Silhouette Score: {silhouette:.4f})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
